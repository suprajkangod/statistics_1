height of learners is following normal dist with a mean of 5.5 ft and standard deviation of 0.25 ft. 
1. what is the probability of the learner's height has less than 5 ft?

2. What is the proabbilty of the leatner's height has more than 5 ft?

3. What is the probability the learner height lies b/w 5.25 to 5.75 ft

4. what is the minimum height of the learner of 60% of the class?

5. what is the maximum height of the learner of 30% of the class?

```python
mu= 5.5
sigma =0.25
```

```python
#1 what is the probability of the learner's height has less than 5 ft? P(x<=5)
st.norm.cdf(5,mu,sigma)
```

```python
st.norm.ppf(0.022750131948179195,mu,sigma)
```

```python
#1 what is the probability of the learner's height has more than 5 ft? P(x>=5)
st.norm.sf(5,mu,sigma)
```

```python
# What is the probability the learner height lies b/w 5.25 to 5.75 ft
# p( 5.25 =< x =<5.75)=?
```

```python
#m1
1- (st.norm.cdf(5.25,mu,sigma)+st.norm.sf(5.75, mu,sigma))
```

```python
#m2  - mu +- 1 sigma  = 68.2%
st.norm.cdf(5.75,mu,sigma)-st.norm.cdf(5.25,mu,sigma)
```

```python
#m3
st.norm.sf(5.25,mu,sigma)-st.norm.sf(5.75,mu,sigma)
```

```python
#mu +- 2 sigma  - 95.4%
st.norm.cdf(6,mu,sigma)-st.norm.cdf(5,mu,sigma)
```

```python
#mu +- 3 sigma  = 99.73%
st.norm.cdf(6.25,mu,sigma)-st.norm.cdf(4.75,mu,sigma)
```

```python
#what is the minimum height of the learner of 60% of the class?
# given is sf= 0.6
st.norm.isf(0.6, mu,sigma)
```

```python
#what is the maximum height of the learner of 30% of the class?
st.norm.ppf(0.3,mu,sigma)
```
